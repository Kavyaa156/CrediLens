Handling Missing Values

Two columns contained missing values: MonthlyIncome and NumberOfDependents.
These were handled using median imputation to preserve dataset size while reducing the impact of outliers.

```python
# Check max value
print("Max Utilization:", df['RevolvingUtilizationOfUnsecuredLines'].max())
```

```python
# Remove extreme outliers in RevolvingUtilizationOfUnsecuredLines
df = df[df['RevolvingUtilizationOfUnsecuredLines'] < 5]

# Check max value
print("Max Utilization:", df['RevolvingUtilizationOfUnsecuredLines'].max())
```

```python
print(df['DebtRatio'].max())
```

```python
df = df[df['DebtRatio'] < 10]

print(df['DebtRatio'].max())
```

```python
# STEP 12: Check if values 96, 97, 98 exist in numeric columns

for column in df.columns:
    if df[column].dtype != 'object':
        count = df[column].isin([96, 97, 98]).sum()
        print(column, ":", count)
```

```python
import numpy as np

late_columns = [
'NumberOfTime30-59DaysPastDueNotWorse',
'NumberOfTime60-89DaysPastDueNotWorse',
'NumberOfTimes90DaysLate'
]

for col in late_columns:
    df[col] = df[col].replace([96, 98], np.nan)
```

```python
for col in late_columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
df["TotalPastDue"] = (
    df["NumberOfTime30-59DaysPastDueNotWorse"]
    + df["NumberOfTime60-89DaysPastDueNotWorse"]
    + df["NumberOfTimes90DaysLate"]
)
```
