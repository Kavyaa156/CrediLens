```python
test_df = test_df[test_df["age"] > 18]
test_df
```

```python
test_df.isnull().sum()
```

```python
import numpy as np

late_columns = [
'NumberOfTime30-59DaysPastDueNotWorse',
'NumberOfTime60-89DaysPastDueNotWorse',
'NumberOfTimes90DaysLate'
]

for col in late_columns:
    test_df[col] = test_df[col].replace([96, 98], np.nan)
```

```python
for col in late_columns:
    test_df[col] = test_df[col].fillna(test_df[col].median())
```

```python
test_df["TotalPastDue"] = (
    test_df["NumberOfTime30-59DaysPastDueNotWorse"]
    + test_df["NumberOfTime60-89DaysPastDueNotWorse"]
    + test_df["NumberOfTimes90DaysLate"]
)
```
